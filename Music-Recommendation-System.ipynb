```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./music.csv')
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
data.head()
```

```python
lst = [] # Taking the point values in the list
for i in data['User-Rating']:
    lst.append(i[:3])
lst
```

```python
data.loc[:,'User-Rating'] = lst
```

```python
data.head()
```

```python
data['Album/Movie'] = data['Album/Movie'].str.replace(' ', '')
data['Singer/Artists'] = data['Singer/Artists'].str.replace(' ','')
data.head()
```

```python
data['Singer/Artists'] = data['Singer/Artists'].str.replace(',',' ')
```

```python
data.head()
```

```python
data['tags'] = data['Singer/Artists'] + ' ' + data['Genre'] + ' '+data['Album/Movie'] + ' ' +data['User-Rating']
data['tags']
```

```python
newData = data[['Song-Name', 'tags']]
newData
```

```python
newData['tags'] = newData['tags'].apply(lambda x: x.lower())
newData
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000)
```

```python
vectors = cv.fit_transform(newData['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(vectors)
```

```python
sim[0]
```

```python
def recommend(music):
    musicIndex = newData[newData['Song-Name'] == music].index[0]
    distances = sim[musicIndex]
    musicList = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in musicList:
        print(newData.iloc[i[0]]['Song-Name'])
```

```python
recommend('Tere Bin')
```
